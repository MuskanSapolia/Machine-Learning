```python
import numpy as np
import pandas as pd
from sklearn.decomposition import TruncatedSVD
```

```python
columns=['user_id','item_id','rating','timestamp']
user = pd.read_csv('/home/muskan/personal/Datasets/ml-100k/u.data', sep='\t',header=None,names=columns)
user.head()
```

```python
columns = ['item_id' ,'movie title',  'release date' , 'video release date','IMDb URL' , 'unknown' , 'Action' , 'Adventure' , 'Animation', "Children's" , 'Comedy' , 'Crime', 'Documentary', 'Drama', 'Fantasy','Film-Noir' , 'Horror'  ,'Musical' , 'Mystery' , 'Romance' , 'Sci-Fi' , 'Thriller', 'War', 'Western' ]
movies = pd.read_csv('/home/muskan/personal/Datasets/ml-100k/u.item',sep='|',names=columns)
movies.head()
```

```python
movie_names = movies[['item_id','movie title']]
data = pd.merge(user,movie_names , on='item_id')
data.head()
```

```python
data.groupby('item_id')['rating'].count().sort_values(ascending=False).head()
```

```python
highest_rated = data.item_id == 50
data[highest_rated]['movie title'].unique()
```

```python
rating_table = data.pivot_table(values='rating',index='user_id',columns='movie title',fill_value=0)
movie_rating = rating_table.T
movie_rating.head()
```

```python
SVD = TruncatedSVD(n_components = 12, random_state=1)
matrix = SVD.fit_transform(movie_rating)
cor = np.corrcoef(matrix)
cor[:5,:5]
```

```python
movie_names = rating_table.columns
movies = list(movie_names)
cor_star_wars = cor[movies.index('Star Wars (1977)')]
list(movie_names[(cor_star_wars<1.0) & (cor_star_wars>0.9)])
```
