```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import mode
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
train = pd.read_csv('/home/muskan/personal/Datasets/BigmartTrain.csv')
test = pd.read_csv('/home/muskan/personal/Datasets/BigmartTest.csv')
```

```python
print train.shape
print test.shape
print train.info()
print test.info()
```

```python
categorical_features = train.select_dtypes(include = [np.object])
numerical_features = train.select_dtypes(include = [np.int, np.float])
print categorical_features.columns
print numerical_features.head(1)
categorical_features_test = test.select_dtypes(include = [np.object])
numerical_features_test = test.select_dtypes(include = [np.int, np.float])
print categorical_features_test.columns
numerical_features_test.head(1)
```

```python
print train.describe()
test.describe()
```

```python
for col in train.columns:
    print col
    print train[col].value_counts()
```

```python
for col in test.columns:
    print col
    print test[col].value_counts()
```

```python
train.Item_Weight = train.Item_Weight.fillna(train.Item_Weight.mean())
print train.info()
test.Item_Weight = test.Item_Weight.fillna(test.Item_Weight.mean())
test.info()
```

```python
outlet_size_mode = train.pivot_table(values='Outlet_Size', columns='Outlet_Type',aggfunc=(lambda x:mode(x.astype('str')).mode[0]))
print outlet_size_mode 
missing_values = train['Outlet_Size'].isnull() 
train.loc[missing_values,'Outlet_Size'] = train.loc[missing_values,'Outlet_Type'].apply(lambda x: outlet_size_mode[x])
print train.info()
```

```python
outlet_size_mode = test.pivot_table(values='Outlet_Size', columns='Outlet_Type',aggfunc=(lambda x:mode(x.astype('str')).mode[0]))
print outlet_size_mode 
missing_values = test['Outlet_Size'].isnull() 
test.loc[missing_values,'Outlet_Size'] = test.loc[missing_values,'Outlet_Type'].apply(lambda x: outlet_size_mode[x])
print test.info()
```

```python
train.Item_Visibility = train.Item_Visibility.replace(0,train.Item_Visibility.mean())
train.Item_Visibility.head()
print train.describe()
test.Item_Visibility = test.Item_Visibility.replace(0,test.Item_Visibility.mean())
test.Item_Visibility.head()
test.describe()
```

```python
train['Item_Fat_Content'] = train['Item_Fat_Content'].replace({'LF':'Low Fat',
                                                             'reg':'Regular',
                                                             'low fat':'Low Fat'})
print train['Item_Fat_Content'].value_counts()
test['Item_Fat_Content'] = test['Item_Fat_Content'].replace({'LF':'Low Fat',
                                                             'reg':'Regular',
                                                             'low fat':'Low Fat'})
test['Item_Fat_Content'].value_counts()
```

```python
le = LabelEncoder()
var_mod = ['Item_Identifier','Item_Fat_Content','Item_Type','Outlet_Location_Type','Outlet_Size','Outlet_Type','Outlet_Identifier']
for i in var_mod:
    train[i] = le.fit_transform(train[i])
for col in categorical_features:
    print col
    print train[col].value_counts()
```

```python
var_mod = ['Item_Identifier','Item_Fat_Content','Item_Type','Outlet_Location_Type','Outlet_Size','Outlet_Type','Outlet_Identifier']
for i in var_mod:
    test[i] = le.fit_transform(test[i])
for col in categorical_features_test:
    print col
    print test[col].value_counts()
```

```python
x = train.iloc[:,0:11]
y = train.iloc[:,11]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3, random_state = 1)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
pd.DataFrame(zip(x.columns,lr.coef_),columns=['features','estimated coef'])
```

```python
print(r2_score(y_test,y_pred))
print(np.sqrt(mean_squared_error(y_test,y_pred)))
```
